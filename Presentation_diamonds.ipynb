```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dia=pd.read_csv('diamonds.csv')
```

```python
dia
```

```python
dia.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
dia
```

```python
def change_cut(cols):
    if cols == "Fair":
        return 0
    elif cols == "Good":
        return 1
    elif cols == "Very Good":
        return 2
    elif cols == "Premium":
        return 3
    else:
        return 4
```

```python
dia["cut"]=dia["cut"].apply(change_cut)
```

```python
def change_color(cols):
    if cols=='J':
        return 0
    elif cols=='I':
        return 1
    elif cols=='H':
        return 2
    elif cols=='G':
        return 3
    elif cols=='F':
        return 4
    elif cols=='E':
        return 5
    else:
        return 6
    
```

```python
dia["color"]=dia["color"].apply(change_color)
```

```python
def change_clarity(cols):
    if cols=='I1':
        return 0
    elif cols=='SI2':
        return 1
    elif cols=='SI1':
        return 2
    elif cols=='VS2':
        return 3
    elif cols=='VS1':
        return 4
    elif cols=='VVS2':
        return 5
    elif cols=='VVS1':
        return 6
    else:
        return 7
```

```python
dia["clarity"]=dia["clarity"].apply(change_clarity)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=dia.drop('carat',axis=1)
y=dia['carat']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=58)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
```

```python
predictions
```

```python
y_test
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
```

```python
sns.distplot((y_test-predictions),bins=100, kde=False);
```

```python
print('MAE: ', metrics.mean_absolute_error(y_test,predictions))
print('MSE: ', metrics.mean_squared_error(y_test,predictions))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```
